# imports

```python
import requests
import pandas as pd
```

# acquire data

```python
api_url = "https://opendata.maryland.gov/api/id/crti-ybyp.json?$select=*&$order=`:id`+ASC&$limit=3000&$offset=0 "
response = requests.get(api_url)
data = response.json()
data[:5]
```

```python
df = pd.DataFrame(data)
```

# clean data

```python
backup_df = df.copy()
```

```python
df['recieved_date'] = pd.to_datetime(df['recieved_date'])
df['incident_date'] = pd.to_datetime(df['incident_date'])
df['incident_closed_date'] = pd.to_datetime(df['incident_closed_date'])
```

# analyze data

```python
last_year_df = df.loc[df.recieved_date>'2023-03-09'].copy()
```

```python
last_year_df.complaint_type.value_counts()
```

```python
last_year_df.loc[last_year_df.complaint_type.str.contains('Air')]
```

```python
last_year_df['complaint_type'] = last_year_df.complaint_type.str.split()
```

```python
complaints_by_type = last_year_df.explode('complaint_type')
```

```python
complaints_by_type.complaint_type.value_counts()
```

```python
last_year_df.county.value_counts()
```

```python
last_year_df[last_year_df.county.isin(['Not Yet Determined', 'Statewide'])].to_csv('exported_data/invalid_counties.csv', index=False)
```

```python
last_year_df = last_year_df[~last_year_df.county.isin(['Not Yet Determined', 'Statewide'])]
```
